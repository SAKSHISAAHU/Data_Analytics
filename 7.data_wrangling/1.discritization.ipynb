# **DISCRITIZATION**
- It is a process of converting continous variable into discrete variables.

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('/Users/sakshisahu/Documents/learning/Data_Analytics/data/Stroke.csv')
df.head()
```

```python
# encoding stroke column in raw data --- data cleaning
df['stroke']=df['stroke'].replace({0:'no',1:'yes'})
df.head()
```

```python
# how many have stroke and not

df['stroke'].value_counts()
```

```python
# avg age of people having stroke or not

df.groupby('stroke')['age'].mean()
# groupby create categorical group of the given column and can use apply operation.
```

##### to findout which age category have more number of stroke problem we need to do **crosstab** and for that we need to discret column 'age' and 'stroke' bur age is continous column. So, in order to do crosstab we need do  **discritization** -- converting continous to discrete column for that we use **pd.cut(df[col_name], bins= ,lables= )**

```python
df['age_catgr']=pd.cut(df['age'], bins=[0,1,3,12,19,30,60,90], labels=['infant','toddler','children','teenager','adult','mid-age','senior'])
df.head()
```

```python
pd.crosstab(df['stroke'],df['age_catgr'],margins=True)
```
