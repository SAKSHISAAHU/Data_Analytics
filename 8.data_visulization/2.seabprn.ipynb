### **1. Histogram**
- **VARIABLE** - 1 continous
- **USE** - distibution of data
- **SYNTAX**
    <pre>
    sn.hisplot(df[continous_col], bins= ,kde=True/False)
    plt.xticks(range=(min, max, interval_size))
    <pre>

```python
plt.figure(figsize=(5,3))
sn.histplot(df['total_bill'],bins=3,kde=True)
plt.xticks(range(0,60,5))
plt.title('bill distribution')
plt.show()
```

### **2. kde Plot**
- **VARIABLE** - 1 continous
- **USE**
    - tells if the data is skewed or nrmly distibuted.
    - used in data exploration step
- **SYNTAX**
    <pre>
    sn.kdeplot(df[continous_col])
    <pre>

```python
plt.figure(figsize=(5,3))
sn.kdeplot(df['total_bill'])
plt.show() # right skewed
```

#### **3. Boxplot**
- **VARIABLE** - 1 continous
- **USE** - represent 5 mumber summary and whether outlier present or not.
- **SYNTAX** 
    <pre>
    sn.boxplot(x/y=df[continous_col])
    <pre>

```python
plt.figure(figsize=(5,3))
sn.boxplot(x=df['tip'])
plt.show() # outlier present
```

### **4. Scatter Plot**
- **VARIABLE** - 2 continous
- **USE** 
    - represent correlation between variables and direction
    - if values are close to each other then highly corelated
    - if values are far away from each other then lessly corelated
- **SYNTAX** 
    1. sn.scatterplot(x=df[1_continous], y=df[2_continous])
    
    2. sn.scatterplot(x='1_continous', y='2_continous', data=df_name)
    3. sn.relplot(x='1_continous', y='2_continous', data=df_name, hue='dicrete_col')
        - hue - gives different color to different categories present in discrete col.
    4. sn.relplot(x='1_continous', y='2_continous', data=df_name, col='discrete_col', col_wrap= no. of columns wants)
        - col - create different plot for different categories of discrete column
        - col_wrap - no. of columns to adjust different categories plots 
   

```python
plt.figure(figsize=(7,3))

plt.subplot(1,2,1)
sn.scatterplot(x=df['total_bill'], y=df['tip']) # syntax 1

plt.subplot(1,2,2)
sn.scatterplot(x='total_bill', y='tip', data=df) # syntax 2
```

```python
sn.relplot(x='total_bill', y='tip', data=df, hue='sex')
plt.show()
```

```python
sn.relplot(x='total_bill', y='tip', data=df, col='sex')
plt.show()
```
