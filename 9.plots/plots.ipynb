# **Plots**
1. 1 C --> marks
2. 1 D --> gender
3. 2 C --> marks + cgpa
4. 2 D --> gender + subject
5. 1 D + 1 C --> gender + marks
6. 1 D + >1 C --> gender + marks + cgpa
7. 1 C + >1 D --> marks + gender + subject
8. more than 1 D + >1 C --> gender + subject + marks + cgpa
9. 1 time-series + 1 C --> date + sales
10. 1 time-series + >1 C --> date + sale + profit
11. 1 time_series + 1 C + 1 D --> date + sale + region
12. 1 time-series + >1 C + 1 D --> date + sale + profit + region

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('sample.csv')
df
```

## **1. 1 Continous**
- histogram
- boxplot
- kde plot
- violine plot

```python
plt.figure(figsize=(10,10))

# histogram
plt.subplot(2,2,1)
plt.hist(df['marks'], bins=4)
plt.xlabel('marks')
plt.ylabel('count')
plt.title('histogram')

# boxplot
plt.subplot(2,2,2)
plt.boxplot(df['marks'])
plt.ylabel('marks')
plt.title('boxplot')

# violine plot
plt.subplot(2,2,3)
sns.violinplot(df['marks'])
plt.title('violine')

# kde plot
plt.subplot(2,2,4)
sns.kdeplot(df['marks'])
plt.title('kde plot')

plt.show()
```

## **2. 1 Discrete**
- count plot
- pie plot

```python
plt.figure(figsize=(7,7))

# count plot
plt.subplot(2,2,1)
sns.countplot(x=df['gender'])
plt.title('count plot')

# pie plot
plt.subplot(2,2,2)
plt.pie(df['gender'].value_counts(), labels=df['gender'].value_counts().index.tolist(), autopct='%0.0f%%')
plt.title('pie plot')

plt.show()
```

## **3. 2 Continous**
- scatter plot
- join plot 
- pair plot 

```python
# scatter plot
plt.scatter(x=df['marks'], y=df['cgpa'])
plt.title('scatter plot')

# join plot
sns.jointplot(x=df['marks'], y=df['cgpa'])

# pair plot
sns.pairplot(data=df, vars=['marks','cgpa'])

plt.show()
```
